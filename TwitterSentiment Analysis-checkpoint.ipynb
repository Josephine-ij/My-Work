```python
def build_freqs(tweets, labels):
    freqs = {}
    for label, tweet in zip(labels, tweets):
        for word in tweet:
            pair = (word, int(label))
            freqs[pair] = freqs.get(pair, 0) + 1
    return freqs
```

```python
Twitter_Data_df['clean_text'].fillna('', inplace=True)
```

```python
freqs = build_freqs(Twitter_Data_df['clean_text'], Twitter_Data_df['category'])
```

```python
word_freq_list = [(word, freqs.get((word, 1), 0), freqs.get((word, 0), 0)) for word in freqs.keys()]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = [entry[1] for entry in word_freq_list]
y = [entry[2] for entry in word_freq_list]
ax.scatter(x, y)
plt.xlabel("Positive count")
plt.ylabel("Negative count")
plt.title("Word Frequency Distribution")
plt.show()
```

```python
for entry in word_freq_list:
    ax.annotate(entry[0], (np.log(entry[1] + 1), np.log(entry[2] + 1)))
```

```python
plt.xlabel("Log Positive count")
plt.ylabel("Log Negative count")
plt.title("Word Frequency in Positive and Negative Tweets")
plt.grid(True)
plt.show()
```

```python
word_freq_dict = {}
for tokens in Twitter_Data_df['clean_text_tokens']:
    for word in tokens:
        word_freq_dict[word] = word_freq_dict.get(word, 0) + 1
```

```python
sorted_word_freq = sorted(word_freq_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
N = 20  
top_words = [entry[0] for entry in sorted_word_freq[:N]]
word_counts = [entry[1] for entry in sorted_word_freq[:N]]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(top_words, word_counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution")
plt.xticks(rotation=45)
plt.show()
```
