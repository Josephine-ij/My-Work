```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
creditcard_df = pd.read_excel('C:\\Users\\o\\Documents\\creditcard.xlsx')
```

```python
creditcard_df
```

```python
features = ['V%d' % number for number in range(1, 29)] + ['Amount']
X_features = creditcard_df[features]
target = creditcard_df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, target, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler
```

```python
anomaly_detector = IsolationForest(contamination=0.01, random_state=42)
anomaly_detector.fit(X_train_scaled)
```

```python
outlier_labels = anomaly_detector.predict(X_test_scaled)
anomalies = X_test_scaled[outlier_labels == -1]
```

```python
for anomaly in anomalies:
    if anomaly_detector.predict([anomaly])[0] == -1:
        print("Potential fraudulent activity detected:", anomaly)
```

```python
logistic_regression_model = LogisticRegression(max_iter=1000)
logistic_regression_model.fit(X_train_scaled, y_train)
y_pred_log_reg = logistic_regression_model.predict(X_test_scaled)
```

```python
print(classification_report(y_test, y_pred_log_reg))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_log_reg))
```

```python
decision_tree_classifier = DecisionTreeClassifier()
decision_tree_classifier.fit(X_train_scaled, y_train)
y_pred_dt = decision_tree_classifier.predict(X_test_scaled)
```

```python
print(classification_report(y_test, y_pred_dt))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_dt))
```

```python
random_forest_classifier = RandomForestClassifier(n_estimators=100)
random_forest_classifier.fit(X_train_scaled, y_train)
y_pred_rf = random_forest_classifier.predict(X_test_scaled)
```

```python
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
plt.figure(figsize=(10, 6))
feat_importances = pd.Series(decision_tree_classifier.feature_importances_, index=features)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importance (Decision Trees)')
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Blues")
plt.title('Confusion Matrix (Random Forest)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
