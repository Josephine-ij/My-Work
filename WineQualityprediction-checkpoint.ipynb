```python
features = WineQT_df.drop(columns="quality")  
target = WineQT_df["quality"] 
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=42)

print("Training Features Shape:", features_train.shape)
print("Testing Features Shape: ", features_test.shape)
print("Training Target Shape:  ", target_train.shape)
print("Testing Target Shape:   ", target_test.shape)
```

```python
lr_model = LinearRegression()

lr_model.fit(features_train, target_train)

print("Training Score (R^2): ", lr_model.score(features_train, target_train))
print("Testing Score (R^2): ", lr_model.score(features_test, target_test))

y_pred_lr = lr_model.predict(features_test)

print("Mean Absolute Error: ", mean_absolute_error(target_test, y_pred_lr))
print("Mean Squared Error: ", mean_squared_error(target_test, y_pred_lr))
print("Median Absolute Error: ", median_absolute_error(target_test, y_pred_lr))
```

```python
logistic_model = LogisticRegression(solver='liblinear')

logistic_model.fit(features_train, target_train)

print("Training Accuracy: ", logistic_model.score(features_train, target_train))
print("Testing Accuracy: ", logistic_model.score(features_test, target_test))

y_pred_logistic = logistic_model.predict(features_test)

print("Mean Absolute Error: ", mean_absolute_error(target_test, y_pred_logistic))
print("Mean Squared Error: ", mean_squared_error(target_test, y_pred_logistic))
print("Median Absolute Error: ", median_absolute_error(target_test, y_pred_logistic))

print("Accuracy Score: ", accuracy_score(target_test, y_pred_logistic))
```

```python
tree_model = DecisionTreeClassifier(max_depth=10)

tree_model.fit(features_train, target_train)

print("Training Accuracy: ", tree_model.score(features_train, target_train))
print("Testing Accuracy: ", tree_model.score(features_test, target_test))
```

```python
print("Feature Importance Scores:")
print(tree_model.feature_importances_)
```
