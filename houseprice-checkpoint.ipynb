```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

Housing_df[varlist] = Housing_df[varlist].apply(binary_map)
```

```python
Housing_df.info()
```

```python
Housing_df.head()
```

```python
status = pd.get_dummies(Housing_df['furnishingstatus'], drop_first=True)
```

```python
status
```

```python
Housing_df = pd.concat([Housing_df, status], axis=1)
```

```python
Housing_df.head()
```

```python
Housing_df.drop(['furnishingstatus'], axis=1, inplace=True)
```

```python
Housing_df.head()
```

```python
np.random.seed(0)
df_train, df_test = train_test_split(Housing_df, train_size=0.7, test_size=0.3, random_state=100)
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
print(df_train.head())
```

```python
df_train.describe()
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df_train.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap of DataFrame")
plt.show()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(estimator=lm, n_features_to_select=6)            
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
X_train_rfe = sm.add_constant(X_train_rfe) 
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
vif_df = pd.DataFrame()
X = X_train_rfe
vif_df['Features'] = X.columns
vif_df['VIF'] = [vif(X.values, i) for i in range(X.shape[1])]
vif_df['VIF'] = round(vif_df['VIF'], 2)
vif_df = vif_df.sort_values(by="VIF", ascending=False)

vif_df
```

```python
y_train_predicted = lm.predict(X_train_rfe)
```

```python
residuals = y_train_predicted - y_train
```

```python
fig = plt.figure(figsize=(8, 6))
sns.distplot(residuals, bins=20)
plt.title('Distribution of Error Terms', fontsize=20)
plt.xlabel('Errors', fontsize=18)
plt.ylabel('Frequency', fontsize=18)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_train, residuals, alpha=0.5, color='blue')
plt.title('Residuals vs. Actual Prices', fontsize=20)
plt.xlabel('Actual Prices', fontsize=18)
plt.ylabel('Residuals', fontsize=18)
```
